## 도로와 가까운 건물 샘플링

```python
# 도로 코드
roads_code = list(roads_15.RN_CD.unique())
len(roads_code)
```

```python
# 샘플링
result = sp.loc[sp.RN_CD.isin(roads_code),:]
result = result.rename(columns = {'빌딩코드': 'BD_MGT_SN'})
result_code = list(result.BD_MGT_SN.unique())
result
```

```python
# 중복값 제거
result_uni = result.drop_duplicates("BD_MGT_SN")
result_uni
```

```python
BB= pd.read_csv('../data/high_bld_over20_include_BIBS.csv', encoding='euc-kr')
Bb = BB.loc[BB.빌딩코드.isin(result_code),['빌딩코드','lat_center','lon_center','Bl','Bs']]
Bb.columns = ['BD_MGT_SN','lat_center','lon_center','BI','Bs']
Bb
```
