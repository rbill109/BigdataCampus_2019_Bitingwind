# 분별로 카운팅(group by 관측소)

```python
# 폴더 안에 있는 데이터 모두 대입해서 데이터 프레임 만들기
from os import listdir
from os.path import isfile, join

files = [f for f in listdir('../data/weather/기상_rawfile_csv/') if isfile(join('../data/weather/기상_rawfile_csv/', f))]
files
        
# 파일 경로명 리스트 생성하기
file_path_list = []
for file in files :
    path = '../data/weather/기상_rawfile_csv/'
    file_path_list.append(path+file)

# file_path_list
```

```python
data_sample = pd.read_csv("../data/weather/기상_rawfile_csv/1903.csv",encoding="euc-kr")

# 전체 데이터 합치기
wind_all = pd.DataFrame(columns=list(data_sample.columns))
for file_path in file_path_list :
    data = pd.read_csv("%s" %file_path, encoding="euc-kr")
    wind_all = pd.concat([wind_all,data],ignore_index=True)
```

```python
# 데이터 확인
wind_all.info()
```

```python
# 풍속 5 넘는 날 세서 분당 카운팅하기
wind_over5 = wind_all[wind_all.wind_speed >= 5]
groups = wind_over5.groupby(wind_over5.station)
wind_cnt = pd.DataFrame(groups["wind_speed"].count())
wind_cnt = wind_cnt.sort_values("wind_speed",ascending=False).reset_index()
wind_cnt.head()
```

```python
wind_loc = pd.read_csv("../data/seoul_station_xy_vals.csv",encoding="euc-kr")
del wind_loc["Unnamed: 0"]
wind_loc = wind_loc.rename(columns={"지점명":"station",
                                   "주소":"address",
                                   "위도":"lat",
                                   "경도":"lon"})
wind_loc
```

```python
from pyproj import Proj, transform
import numpy as np
import pandas as pd

proj_EPSG = Proj(init='epsg:5186') # 한국형 좌표계
proj_WGS = Proj(init='epsg:4326') # GPS사용 전지구 좌표

lat_list = []
lon_list = []
for n in range(0,len(wind_loc.index)) :
    try :
        lat_put = wind_loc.lat[n]
        lon_put = wind_loc.lon[n]
        lon_new, lat_new  = pyproj.transform(proj_WGS, proj_EPSG, lon_put, lat_put)
        lat_list.append(lat_new)
        lon_list.append(lon_new)
    except :
        lat_list.append(np.nan)
        lon_list.append(np.nan)
        
wind_loc["lat_EPSG"] = lat_list
wind_loc["lon_EPSG"] = lon_list

wind_loc = wind_loc.rename(columns = {"lat":"lat_WGS","lon":"lon_WGS"})
```

```python
wind_high_cnt = pd.merge(wind_cnt,wind_loc)
wind_high_cnt.to_csv("../data/weather/wind_high_cnt.csv",encoding="euc-kr")
```

```python
def day_counter(w_data) :
    cnt_vector = ""
    for i in w_data.wind_speed :
        if i >= 5 :
            cnt_vector += "1"
        else :
            cnt_vector += "0"
    real = cnt_vector

    cnt = []
    for i in range(0,len(cnt_vector)) :
        if cnt_vector[i] == "0" :
            pass
        else :
            try :
                if cnt_vector[i+1] == "0" :
                    pass
                else :
                    n = len(cnt_vector[i:].split("0")[0])
                    cnt.append(n)
                    cnt_vector[i:i+n-1].replace("1","0")
            except IndexError :
                    for i in range(0,len(cnt_vector)-1) :
                        if cnt_vector[i] == "0" :
                            pass
                        else :
                            if cnt_vector[i+1] == "0" :
                                pass
                            else :
                                n = len(cnt_vector.split("0")[-1])
                                cnt.append(n)

    cnts_over5 = []
    for n in cnt :
        if n >=5 :
            cnts_over5.append(n)
        else :
            pass
        
    return real, cnts_over5
```
