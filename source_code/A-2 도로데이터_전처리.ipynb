# 대로만 뽑아보기

```python
import geopandas as gpd
import pandas as pd
```

```python
data = pd.read_excel("../data/서울시_도로구간_위치정보.xlsx")
data.head()
```

```python
data_2 = gpd.read_file("../data/2014_building/TL_SPRD_MANAGE_W.shp")
data_2.head()
```

```python
print("엑셀파일 길이 :",len(data))
print("shape파일 길이 :",len(data_2))
# shp 파일로 전처리 결정
```

```python
# '대로' 포함한 길 추출
data_대로 = data_2[data_2.RN.str.contains("대로")==True].drop_duplicates()

# '길' 포함한 길 제거
data_대로_real = data_대로[data_대로.RN.str.contains("길")==False]
data_대로_real

data_대로_real
```

```python
# 저장하기
data_대로_real.to_file("../data/2014_building/real_big_roads.shp",encoding="euc-kr")
```
